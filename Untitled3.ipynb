```python
import pandas as pd

# Load the dataset with the correct file name
df = pd.read_csv('data/Reviews.csv')

# Select the review text column ('Text') for sentiment analysis
df = df[['Text']]  # Using 'Text' as the column for reviews
df.head()
```

```python
import nltk
import re
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def preprocess_text(text): # Change parameter name from 'Text' to 'text' to match the function's internal logic
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = text.lower()  # Convert to lowercase
    text = ' '.join(word for word in text.split() if word not in stop_words)
    return text

df['cleaned_text'] = df['Text'].apply(preprocess_text) # Change 'review_text' to 'Text' to access the correct column in the DataFrame
df.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
!pip install vaderSentiment

analyzer = SentimentIntensityAnalyzer()

def vader_sentiment(text):
    scores = analyzer.polarity_scores(text)
    return scores['compound']  # Use compound score for overall sentiment

df['vader_sentiment'] = df['cleaned_text'].apply(vader_sentiment)
df.head()
```

```python
from textblob import TextBlob
!pip install textblob

def textblob_sentiment(text):
    return TextBlob(text).sentiment.polarity

df['textblob_sentiment'] = df['cleaned_text'].apply(textblob_sentiment)
df.head()
```
